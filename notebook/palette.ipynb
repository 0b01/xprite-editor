# Shrink pixels

Pixel art is usually exported to 2x, 3x, 5x its original size. This is because many most image viewer(including browser) uses antialiasing so the original size artwork would look blurred. The following algorithm shrinks a 2x pixel art image down to 1x.

```python
%matplotlib inline

from PIL import Image
from collections import defaultdict
import numpy as np

import colorsys

from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

# Color distance

Due to jpeg artifacts, many individual pixels are slightly altered so we will need a color distance function to allow for an error.

```python
def color_dist(rgb1,rgb2):
    """https://stackoverflow.com/questions/8863810/python-find-similar-colors-best-way"""
    rm = 0.5*(rgb1[0]+rgb2[0])
    d = sum((2+rm,4,3-rm) * (rgb1-rgb2)**2)**0.5
    return d

color_dist(np.array([1,0,0]), np.array([0,0,0]))
```

# Find pixel width

The idea is to iterate over each row and find the greatest common divisor of top n most frequent lengths of color bands.

```python
def find_pixel_width(im):
    width = defaultdict(lambda: 0)
    (w, h) = im.size
    for col in range(h):
        acc = 1
        oldcolor = np.array(im.getpixel((0, col)))
        for row in range(w):
            color = im.getpixel((row, col))
            color = np.array(color)
            if color_dist(color, oldcolor) > 10.:
                width[acc] += 1
                oldcolor = color
                acc = 1
            else:
                acc += 1
        width[acc] += 1
    return width
```

```python
width_dict = find_pixel_width(im)
d = sorted(width_dict.items(), key=lambda kv: -kv[1])[:10]
d
```

```python
def gcd_multi(l):
    def gcd(x, y): 
        while(y): 
            x, y = y, x % y 
        return x
    num1 = l[0]
    num2 = l[1]
    acc = gcd(num1, num2)
    for i in range(2, len(l)):
        acc = gcd(acc, l[i])
    return acc
assert gcd_multi([2,4,6]) == 2
assert gcd_multi([3,6,9]) == 3
```

```python
pixel_width = gcd_multi([i[0] for i in d])
```

```python
(w, h) = im.size
out = Image.new("RGB", size=(w//pixel_width, h//pixel_width))
```

```python
for i in range(0, w, pixel_width):
    for j in range(0, h, pixel_width):
        col = im.getpixel((i,j))
        out.putpixel((i//pixel_width, j//pixel_width), col)
```

So the linear predictor learns a plane which is not what we want. So after googling around I found an example using skimage: http://scikit-image.org/docs/0.13.x/auto_examples/transform/plot_ransac3D.html

```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from skimage.measure import LineModelND, ransac
```

```python
xyz = np.stack((x, y, V), 1)

# robustly fit line only using inlier data with RANSAC algorithm
model_robust, inliers = ransac(xyz, LineModelND, min_samples=2,
                               residual_threshold=0.09, max_trials=10000)
outliers = inliers == False

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xyz[inliers][:, 0], xyz[inliers][:, 1], xyz[inliers][:, 2], c=c[inliers], s=s[inliers]/10, label='Inlier data')
ax.scatter(xyz[outliers][:, 0], xyz[outliers][:, 1], xyz[outliers][:, 2], c=c[outliers], s=s[outliers]/100, label='Outlier data')
ax.legend(loc='lower left')
plt.show()
```
